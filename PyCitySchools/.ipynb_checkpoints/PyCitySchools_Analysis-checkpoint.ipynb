Analysis
This challenge took data from two sets of data.  The first included math and reading grades for all high school student data for the entire school district, both district schools and charter schools.  The second set included size and budget for each school.  The goal of this analysis is to determine if there are any correlations between the school characteristics and the math and reading scores.  

Of the close to 40,000 students in the district, 75% are passing math and 86% are passing reading.  But only 65% are passing both.  On average, math scores are about 3 points lower than reading scores.  The total budget for the 15 schools is \\$24.6M. Holden HS, a charter school, has the lowest budget at $248k and Bailey HS, a district school, has the highest budget at \\$3.124M.  However, Bailey is a much bigger school than Holden and per capita budgets will be explored later.  

The top eight perfroming schools are all charter schools while the bottom seven performing schools are all district schools.  The lowest performing school, Rodriguez HS has a passing rate of 53\%, while the top performaing school, Carbrerra HS has a passing rate of 92\%.  While their reading passing rates are relatively close (between 80-84\%), their math scores are the biggest difference with passing scores of 66\% versus 94\%.  

Not much has changed over the last four years since both the math and reading scores are consistently within 1-2 percentage points over the four classes within each school.  

Per capita budgets range from \\$578-\\$655 showing that funding is fairly evenly distributed among schools.  However, a closer look shows that schools with the lowest amount of funding have the highest test scores and passing rates, with scores getting worse as school budgets increase.  Passing rates do not appear to be a function of spending and the opposite appears true.  

Scores stay consistent with small (less than 1000 students) and medium (1000 to 2000 students), however they tend to fall off for large schools (2000 to 5000 students).  Particularly with overall passing rates that drop from 90% to 58% from small/medium to large schools.  

The results from this challenge show that charter schools significantly outperform district schools by about 27, 16, 37 percentage points for math, reading, and overall passing rates, respectively. Charter school status appears to be the biggest indicator of high performing schools with small to medium schools also performing better, but to a lesser degree. 

See detailed results below.  

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import pandas as pd

#Apply global formatting to all floats to only show hundredths
pd.options.display.float_format = '{:,.2f}'.format

#Load school and student data sets
schooldata_load = "Resources/schools_complete.csv"
studentdata_load = "Resources/students_complete.csv"

#Read data into pandas DF
schooldata = pd.read_csv(schooldata_load)
studentdata = pd.read_csv(studentdata_load)

#Combine data into a single file; combine on school_name
school_data_full = pd.merge(studentdata, schooldata, how="left", on=["school_name", "school_name"])
school_data_full.head()
```

## District Summary

```python
# Number of unique schools
total_schools = school_data_full["school_name"].nunique()
print("Total number of schools = ", total_schools)

# Total number of students
total_students = school_data_full["student_name"].count()
print("Total number of students = ", total_students)

# Total budget
total_budget = schooldata["budget"].sum()
print("Total budget for area = ", total_budget)

# Average math score
avg_math_score = school_data_full["math_score"].mean()
print("Average math score = ", avg_math_score)

# Average (mean) reading score
avg_reading_score = school_data_full["reading_score"].mean()
print("Average reading score = ", avg_reading_score)

# Calculate the percentage of students who passed math (greather than or equal to 70)
passing_math_count = school_data_full[(school_data_full["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(total_students) * 100
print("Percentage of students passing math = ", passing_math_percentage, "%")

# Calculate the percentage of students who passed reading (greather than or equal to 70)
passing_reading_count = school_data_full[(school_data_full["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(total_students) * 100
print("Percentage of students passing reading = ", passing_reading_percentage, "%")

# Calculate the percentage of students that passed both math and reading
passing_math_and_reading_count = school_data_full[
    (school_data_full["math_score"] >= 70) & (school_data_full["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_and_reading_count /  float(total_students) * 100
print("Percentage of students passing math and reading = ", overall_passing_rate, "%")
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame(
{'Total Schools' : [total_schools],
'Total Students' : [total_students],
'Total Budget' : [total_budget],
'Avg Math Score' : [avg_math_score],
'Avg Reading Score' : [avg_reading_score],
'Math Passing %' : [passing_math_percentage],
'Reading Passing %' : [passing_reading_percentage],
'Overall Passing %' : [overall_passing_rate]})

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Avg Math Score"] = district_summary["Avg Math Score"].map("{:,.2f}".format)
district_summary["Avg Reading Score"] = district_summary["Avg Reading Score"].map("{:,.2f}".format)
district_summary["Math Passing %"] = district_summary["Math Passing %"].map("{:,.2f}%".format)
district_summary["Reading Passing %"] = district_summary["Reading Passing %"].map("{:,.2f}%".format)
district_summary["Overall Passing %"] = district_summary["Overall Passing %"].map("{:,.2f}%".format)


# Display the DataFrame
district_summary.style.hide_index()
```

### SCHOOL SUMMARY

```python
# Select the school type
school_type = schooldata.set_index(["school_name"])["type"]

# Calculate the total student count
per_school_counts = school_data_full.groupby(["school_name"]).count()["student_name"]

# Calculate the total school budget and per capita spending
per_school_budget = school_data_full.groupby(["school_name"]).mean()["budget"]
per_student_budget = per_school_budget / per_school_counts

# Calculate the average test scores
per_school_math = school_data_full.groupby(["school_name"]).mean()["math_score"]
per_school_reading = school_data_full.groupby(["school_name"]).mean()["reading_score"]

# Calculate the number of schools with math scores of 70 or higher
school_passing_math = per_school_math[per_school_math >= 70].count()

# Calculate the number of schools with reading scores of 70 or higher
school_passing_reading = per_school_reading[per_school_reading >= 70].count()

# Combine school math and reading schools into one dataframe
school_resultsDF = pd.concat([per_school_math,per_school_reading],axis=1)

# Calculate the schools that passed both math and reading with scores of 70 or higher
passing_math_and_reading = sum((school_resultsDF["math_score"] >= 70) & (school_resultsDF["reading_score"] >= 70))

# Calculate the passing rates
passing_math_count_by_school = school_data_full.groupby('school_name')['math_score'].apply(lambda x: (x>= 70).sum())
per_school_passing_math = passing_math_count_by_school/per_school_counts * 100

passing_reading_count_by_school = school_data_full.groupby('school_name')['reading_score'].apply(lambda x: (x>= 70).sum())
per_school_passing_reading = passing_reading_count_by_school/per_school_counts * 100

passing_math_and_reading = school_data_full[
    (school_data_full["reading_score"] >= 70) & (school_data_full["math_score"] >= 70)
]
overall_passing_rate = passing_math_and_reading.groupby(["school_name"]).count()["student_name"] / per_school_counts * 100
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.

# Rename data series name to get proper header
school_type.name="School Type"
per_school_counts.name="Total Students"
per_school_budget.name="Total School Budget"
per_student_budget.name="Per Student Budget"
per_school_math.name="Avg Math Score"
per_school_reading.name="Avg Reading Score"
per_school_passing_math.name="% Passing Math"
per_school_passing_reading.name="% Passing Reading"
overall_passing_rate.name="% Overall Passing"

# Concatonate data series
per_school_summary = pd.concat([school_type, per_school_counts, per_school_budget,per_student_budget, 
                                per_school_math, per_school_reading, per_school_passing_math, per_school_passing_reading,
                                overall_passing_rate
                               ], axis=1)

# Create a copy of the school summary to retain data types before formatting for use later
per_school_summary_formatted = per_school_summary.copy()

# Formatting
per_school_summary_formatted["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary_formatted["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)
per_school_summary_formatted["Avg Math Score"] = per_school_summary["Avg Math Score"].map("{:,.1f}".format)
per_school_summary_formatted["Avg Reading Score"] = per_school_summary["Avg Reading Score"].map("{:,.1f}".format)
per_school_summary_formatted["% Passing Math"] = per_school_summary["% Passing Math"].map("{:,.1f}%".format)
per_school_summary_formatted["% Passing Reading"] = per_school_summary["% Passing Reading"].map("{:,.1f}%".format)
per_school_summary_formatted["% Overall Passing"] = per_school_summary["% Overall Passing"].map("{:,.1f}%".format)

# Display the DataFrame
per_school_summary_formatted
```

## Math Scores by Grade

```python
# Separate the data by grade
ninth_graders = school_data_full[(school_data_full["grade"] == "9th")]
tenth_graders = school_data_full[(school_data_full["grade"] == "10th")]
eleventh_graders = school_data_full[(school_data_full["grade"] == "11th")]
twelfth_graders = school_data_full[(school_data_full["grade"] == "12th")]

# Group by "school_name" and take the mean of each.
ninth_graders_scores = ninth_graders.groupby(["school_name"]).mean()
tenth_graders_scores = tenth_graders.groupby(["school_name"]).mean()
eleventh_graders_scores = eleventh_graders.groupby(["school_name"]).mean()
twelfth_graders_scores = twelfth_graders.groupby(["school_name"]).mean()

# Use the code to select only the `math_score`.
ninth_grade_math_scores = ninth_graders_scores["math_score"]
tenth_grade_math_scores = tenth_graders_scores["math_score"]
eleventh_grade_math_scores = eleventh_graders_scores["math_score"]
twelfth_grade_math_scores = twelfth_graders_scores["math_score"]

# Rename data series name to get proper header
ninth_grade_math_scores.name="9th"
tenth_grade_math_scores.name="10th"
eleventh_grade_math_scores.name="11th"
twelfth_grade_math_scores.name="12th"


# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.concat([ninth_grade_math_scores, tenth_grade_math_scores,  
                               eleventh_grade_math_scores, twelfth_grade_math_scores], axis=1)


# Minor data wrangling
math_scores_by_grade.index.name = None

# Display the DataFrame
print("Math Scores by Grade")
math_scores_by_grade
```

## Reading Scores by Grade

```python
# Use the code to select only the `reading_score`.
ninth_grade_reading_scores = ninth_graders_scores["reading_score"]
tenth_grade_reading_scores = tenth_graders_scores["reading_score"]
eleventh_grade_reading_scores = eleventh_graders_scores["reading_score"]
twelfth_grade_reading_scores = twelfth_graders_scores["reading_score"]

# Rename data series name to get proper header
ninth_grade_reading_scores.name="9th"
tenth_grade_reading_scores.name="10th"
eleventh_grade_reading_scores.name="11th"
twelfth_grade_reading_scores.name="12th"


# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
reading_scores_by_grade = pd.concat([ninth_grade_reading_scores, tenth_grade_reading_scores,  
                               eleventh_grade_reading_scores, twelfth_grade_reading_scores], axis=1)


# Minor data wrangling
reading_scores_by_grade.index.name = None

# Display the DataFrame
print("Reading Scores by Grade")
reading_scores_by_grade
```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]

# make a copy of the clean school_spending_summary
school_spending_df = per_school_summary.copy()

# Use `pd.cut` to categorize spending based on the bins
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(
    school_spending_df["Per Student Budget"], bins = spending_bins,
    labels=spending_labels, right=False)

school_spending_df=school_spending_df.sort_values("school_name")
school_spending_df
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Avg Math Score"]
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Avg Reading Score"]
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]

# Assemble into DataFrame
spending_summary = pd.concat([spending_math_scores, spending_reading_scores, spending_passing_math, 
                               spending_passing_reading, overall_passing_spending], axis=1)


# Display results
spending_summary
```

## Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

# make a copy of the clean school_spending_summary
school_spending_school_size = per_school_summary.copy()

school_spending_school_size["School Size"] = pd.cut(
    school_spending_df["Total Students"], bins = size_bins,
    labels=size_labels, right=False)

school_spending_school_size=school_spending_school_size.sort_values("school_name")
school_spending_school_size
```

```python
# Calculate averages for the desired columns. 
size_math_scores = school_spending_school_size.groupby(["School Size"]).mean()["Avg Math Score"]
size_reading_scores = school_spending_school_size.groupby(["School Size"]).mean()["Avg Reading Score"]
size_passing_math = school_spending_school_size.groupby(["School Size"]).mean()["% Passing Math"]
size_passing_reading = school_spending_school_size.groupby(["School Size"]).mean()["% Passing Reading"]
size_overall_passing = school_spending_school_size.groupby(["School Size"]).mean()["% Overall Passing"]

# Assemble into DataFrame
size_summary = pd.concat([size_math_scores, size_reading_scores, size_passing_math, 
                               size_passing_reading, size_overall_passing], axis=1)


# Display results
size_summary
```
